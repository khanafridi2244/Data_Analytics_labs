```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('school_data.csv')
df.head()
```

```python
# for the frequency count
sns.countplot(df, x="gender")
pd.value_counts(df['gender'])
```

```python
# when same data is needed in a percentage form
df['gender'].value_counts().plot(kind = 'pie', autopct = '%.2f')
```

```python
# histogram as hist use to create a range or distribution
import matplotlib.pyplot as plt

plt.hist(df['ssc_p'],bins=5)
```

```python
# Displot => improve in histogram and tell the probability 
sns.distplot(df['ssc_p'])
```

```python
# Boxplot => identify the outliers
sns.boxplot(df['ssc_p'])
```
