```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset (available in seaborn)
data = sns.load_dataset('titanic')
data.head()
```

## Seaborn
Statistical analysis is a process of understanding how variables in a dataset relate to each other and how 
those relationships depend on other variables. Visualization can be a core component of this process 
because, when data are visualized properly, the human visual system can see trends and patterns that 
indicate a relationship

# Scatter Plot with Hue, Style, and Size

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x="age", y="fare", hue="sex", style="who", size="pclass", sizes=(20, 200))
plt.title('Scatter Plot: Age vs Fare')
plt.show()
```

# Line Plot

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x="age", y="fare", hue="class", style="sex", markers=True)
plt.title('Line Plot: Age vs Fare')
plt.show()
```

# Subplots

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
sns.histplot(data=data, x="age", kde=True, ax=axes[0], color="skyblue")
axes[0].set_title('Age Distribution')
sns.countplot(data=data, x="class", hue="sex", ax=axes[1])
axes[1].set_title('Passenger Class by Gender')
plt.tight_layout()
plt.show()
```

# Cat Plot

```python
sns.catplot(data=data, x="sex", y="age", hue="class", kind="swarm", height=6, aspect=1.5)
plt.title('Categorical Plot: Age by Gender and Class')
plt.show()
```

# Box Plot

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x="pclass", y="fare", hue="sex")
plt.title('Box Plot: Fare by Passenger Class and Gender')
plt.show()
```


# Boxen Plot

```python
plt.figure(figsize=(10, 6))
sns.boxenplot(data=data, x="class", y="age", hue="who")
plt.title('Boxen Plot: Age by Class and Who')
plt.show()
```

# Violin Plot

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=data, x="embark_town", y="fare", hue="sex", split=True)
plt.title('Violin Plot: Fare by Embark Town and Gender')
plt.show()
```

# Bar Plot

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x="class", y="survived", hue="sex")
plt.title('Bar Plot: Survival Rate by Class and Gender')
plt.show()
```

# Pair Plot

```python
sns.pairplot(data=data, hue="sex", diag_kind="kde")
plt.suptitle('Pair Plot: Titanic Dataset', y=1.02)
plt.show()
```

# Regression Plot

```python
plt.figure(figsize=(10, 6))
sns.regplot(data=data, x="age", y="fare", scatter_kws={"alpha": 0.5}, line_kws={"color": "red"})
plt.title('Regression Plot: Age vs Fare')
plt.show()
```

# Controlling Plotted Figure Aesthetics

```python
sns.set_context("talk", font_scale=1.2)
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x="fare", kde=True, color="purple")
plt.title('Adjusted Aesthetics: Fare Distribution')
plt.show()
```

## Plotly & Cufflinks
Plotly is an open-source and browser-based graphing library which facilitates interactive 
plotting. The library is available for a number of programming languages such as Python, R, 
MATLAB, Perl, Julia, Arduino, and REST, among others.
Cufflinks is another library that connects the Pandas data frame with Plotly enabling users to create 
visualizations directly from Pandas. The library binds the power of Plotly with the flexibility of 
Pandas for easy plotting.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Line Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x="age", y="fare", hue="class", style="sex", markers=True)
plt.title('Line Plot: Age vs Fare')
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x="age", y="fare", hue="sex", style="who", size="pclass", sizes=(20, 200))
plt.title('Scatter Plot: Age vs Fare')
plt.show()
```

```python
# Bar Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x="class", y="survived", hue="sex")
plt.title('Bar Plot: Survival Rate by Class and Gender')
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x="pclass", y="fare", hue="sex")
plt.title('Box Plot: Fare by Passenger Class and Gender')
plt.show()
```

```python
# Area Plot
plt.figure(figsize=(10, 6))
data_sorted = data.sort_values(by="age")
plt.fill_between(data_sorted["age"], data_sorted["fare"].fillna(0), alpha=0.5, label="Fare")
plt.title('Area Plot: Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend()
plt.show()
```

```python
# 3D Plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data["age"], data["fare"], data["pclass"], c=data["pclass"], cmap="viridis", s=50)
ax.set_xlabel('Age')
ax.set_ylabel('Fare')
ax.set_zlabel('Pclass')
plt.title('3D Scatter Plot: Age, Fare, Pclass')
plt.show()
```

```python
# Spread Plot (Scatter Plot with Jitter)
data_jitter = data.copy()
data_jitter["age"] += np.random.normal(0, 1, size=len(data))
data_jitter["fare"] += np.random.normal(0, 5, size=len(data))
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data_jitter, x="age", y="fare", hue="sex")
plt.title('Spread Plot: Age vs Fare with Jitter')
plt.show()
```

```python
# Histogram Plot
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x="age", kde=True, color="skyblue")
plt.title('Histogram: Age Distribution')
plt.show()
```

```python
# Bubble Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x="age", y="fare", size="fare", hue="sex", sizes=(20, 300), alpha=0.6, edgecolor="k")
plt.title('Bubble Plot: Age vs Fare')
plt.show()
```

## Time Series Plots

```python
# Line Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x="age", y="fare", hue="class", style="sex", markers=True)
plt.title('Line Plot: Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x="age", y="fare", hue="sex", style="who", size="pclass", sizes=(20, 200))
plt.title('Scatter Plot: Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
# Subplots: Histogram and KDE Plot
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
sns.histplot(data=data, x="age", kde=True, ax=axes[0], color="skyblue")
axes[0].set_title('Histogram: Age Distribution')
sns.kdeplot(data=data, x="fare", hue="class", fill=True, ax=axes[1])
axes[1].set_title('KDE Plot: Fare by Class')
plt.tight_layout()
plt.show()
```

```python

# Histogram Plot
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x="fare", bins=20, kde=True, color="green")
plt.title('Histogram: Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```
