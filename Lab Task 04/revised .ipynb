```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load the tips dataset
df = sns.load_dataset('tips')

# Display basic information
print(df.head())
```

## line plot

```python
# Line Plot: Total Bill over Index
plt.figure(figsize=(8, 5))
plt.plot(df.index, df['total_bill'], label="Total Bill")
plt.xlabel('Index')
plt.ylabel('Total Bill ($)')
plt.title('Line Plot of Total Bill')
plt.legend()
plt.grid()
plt.show()
```

## Bar Plot

```python
# Bar Plot: Average Tip by Day
avg_tip = df.groupby('day')['tip'].mean()
avg_tip.plot(kind='bar', color='skyblue', figsize=(8, 5))
plt.title('Average Tip by Day')
plt.xlabel('Day')
plt.ylabel('Tip ($)')
plt.grid(axis='y')
plt.show()
```

## Stacked Plot

```python
# Stacked Plot: Total Bill and Tip
plt.figure(figsize=(8, 5))
plt.stackplot(df.index, df['total_bill'], df['tip'], labels=['Total Bill', 'Tip'], alpha=0.8)
plt.title('Stacked Plot of Total Bill and Tip')
plt.xlabel('Index')
plt.ylabel('Amount ($)')
plt.legend(loc='upper left')
plt.show()
```

## Histogram

```python
# Histogram: Distribution of Total Bill
plt.figure(figsize=(8, 5))
plt.hist(df['total_bill'], bins=10, color='orange', edgecolor='black', alpha=0.7)
plt.title('Histogram of Total Bill')
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')
plt.grid()
plt.show()

```

## Box Plot

```python
# Box Plot: Total Bill
plt.figure(figsize=(8, 5))
plt.boxplot(df['total_bill'], patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.title('Box Plot of Total Bill')
plt.ylabel('Total Bill ($)')
plt.show()
```

##  Area Plot

```python
# Area Plot: Total Bill and Tip
plt.figure(figsize=(8, 5))
plt.fill_between(df.index, df['total_bill'], color='lightblue', alpha=0.5, label='Total Bill')
plt.fill_between(df.index, df['tip'], color='orange', alpha=0.5, label='Tip')
plt.title('Area Plot of Total Bill and Tip')
plt.xlabel('Index')
plt.ylabel('Amount ($)')
plt.legend()
plt.show()
```

##  Scatter Plot

```python
# Scatter Plot: Total Bill vs Tip
plt.figure(figsize=(8, 5))
plt.scatter(df['total_bill'], df['tip'], color='purple', alpha=0.7)
plt.title('Scatter Plot of Total Bill vs Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.grid()
plt.show()
```

## Hexbin Plot

```python
# Hexbin Plot: Total Bill vs Tip
plt.figure(figsize=(8, 5))
plt.hexbin(df['total_bill'], df['tip'], gridsize=20, cmap='Blues', mincnt=1)
plt.colorbar(label='Count')
plt.title('Hexbin Plot of Total Bill vs Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.show()
```

##  Pie Chart

```python
# Pie Plot: Gender Distribution
gender_counts = df['sex'].value_counts()
plt.figure(figsize=(8, 5))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=['pink', 'lightblue'])
plt.title('Pie Chart of Gender Distribution')
plt.show()
```

##  Donut Chart

```python
# Donut Chart: Gender Distribution
plt.figure(figsize=(8, 5))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=['pink', 'lightblue'])
center_circle = plt.Circle((0, 0), 0.70, color='white')
plt.gca().add_artist(center_circle)
plt.title('Donut Chart of Gender Distribution')
plt.show()
```

##  Scatter Matrix

```python
from pandas.plotting import scatter_matrix

# Scatter Matrix
plt.figure(figsize=(10, 8))
scatter_matrix(df[['total_bill', 'tip', 'size']], alpha=0.8, figsize=(10, 10), diagonal='hist', color='teal')
plt.suptitle('Scatter Matrix of Tips Dataset')
plt.show()
```

## Subplots

```python
# Subplots: Total Bill, Tip, and Size
plt.figure(figsize=(12, 8))

# Subplot 1: Total Bill
plt.subplot(3, 1, 1)
plt.plot(df['total_bill'], color='blue')
plt.title('Total Bill')

# Subplot 2: Tip
plt.subplot(3, 1, 2)
plt.plot(df['tip'], color='green')
plt.title('Tip')

# Subplot 3: Size
plt.subplot(3, 1, 3)
plt.plot(df['size'], color='red')
plt.title('Size')

plt.tight_layout()
plt.show()
```

##  Polar Axis 

```python
# Polar Bar Plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, polar=True)
angles = np.linspace(0, 2 * np.pi, len(gender_counts), endpoint=False).tolist()
values = gender_counts.values
angles += angles[:1]  # Closing the plot
values = np.append(values, values[0])

ax.plot(angles, values, 'o-', linewidth=2)
ax.fill(angles, values, alpha=0.25)
ax.set_yticks([5, 10, 15])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(gender_counts.index)
plt.title("Polar Bar Plot")
plt.show()
```

 ##  Animation Plot

```python
import matplotlib.animation as animation

# Example Animation: Scatter
fig, ax = plt.subplots()
x = df['total_bill']
y = df['tip']
scatter = ax.scatter([], [])

def animate(i):
    ax.clear()
    ax.scatter(x[:i], y[:i], c='blue', alpha=0.7)
    ax.set_xlim(0, x.max())
    ax.set_ylim(0, y.max())
    ax.set_title("Scatter Animation")
    ax.set_xlabel("Total Bill")
    ax.set_ylabel("Tip")

ani = animation.FuncAnimation(fig, animate, frames=len(x), interval=100)
plt.show()
```

##  1. Polar Plot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Load the tips dataset
tips = sns.load_dataset('tips')

# Polar Plot
angles = np.linspace(0, 2 * np.pi, len(tips), endpoint=False)
radii = tips['tip'] / tips['total_bill']  # Tip percentage
plt.figure(figsize=(8, 6))
ax = plt.subplot(111, polar=True)
ax.plot(angles, radii, 'o-', label='Tip %')
ax.fill(angles, radii, alpha=0.25)
ax.legend()
plt.title("Polar Plot of Tip Percentage")
plt.show()
```

## 2. Line Plot on a Polar Axis

```python
angles = np.linspace(0, 2 * np.pi, len(tips), endpoint=False)
radii = tips['total_bill']  # Use total bill for radii
plt.figure(figsize=(8, 6))
ax = plt.subplot(111, polar=True)
ax.plot(angles, radii, label="Total Bill")
ax.legend()
plt.title("Line Plot on Polar Axis")
plt.show()
```

## 3. Scatter Plot on a Polar Axis

```python
plt.figure(figsize=(8, 6))
ax = plt.subplot(111, polar=True)
ax.scatter(angles, radii, c=tips['size'], cmap='viridis', label="Scatter")
plt.colorbar(ax.collections[0], label='Party Size')
plt.title("Scatter Plot on a Polar Axis")
plt.legend()
plt.show()
```

##  4. Integral in Calculus: Plot as Area Under the Curve

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(tips['total_bill'], fill=True, color='skyblue', alpha=0.4)
plt.title("Area Under the Curve: Total Bill Distribution")
plt.xlabel("Total Bill")
plt.ylabel("Density")
plt.show()
```

##  5. Animation Plot

```python
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = plt.plot([], [], 'ro-', animated=True)

def init():
    ax.set_xlim(0, len(tips))
    ax.set_ylim(0, tips['total_bill'].max())
    return ln,

def update(frame):
    xdata.append(frame)
    ydata.append(tips['total_bill'].iloc[frame])
    ln.set_data(xdata, ydata)
    return ln,

ani = FuncAnimation(fig, update, frames=range(len(tips)), init_func=init, blit=True)
plt.title("Animation: Total Bill Over Time")
plt.show()
```

## 5. Waffle Chart

```python
from pywaffle import Waffle

tip_counts = tips['size'].value_counts()
plt.figure(
    FigureClass=Waffle,
    rows=5,
    values=tip_counts,
    title={'label': 'Party Size Distribution', 'loc': 'center'},
    legend={'loc': 'upper left', 'bbox_to_anchor': (1, 1)}
)
plt.show()
```
