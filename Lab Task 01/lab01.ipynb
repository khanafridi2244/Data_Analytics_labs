```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)  # 100 random samples for feature
y = 4 + 3 * X + np.random.randn(100, 1)  # Linear relationship with noise

# Gradient Descent Function
def gradient_descent(X, y, learning_rate=0.01, iterations=1000):
    # Initialize parameters
    m, b = 0, 0  # m = slope, b = intercept
    n = len(y)  # Number of data points
    cost_history = []  # Store cost values

    for _ in range(iterations):
        # Predict y
        y_pred = m * X + b
        
        # Compute gradients
        dm = -(2/n) * np.sum(X * (y - y_pred))  # Partial derivative w.r.t m
        db = -(2/n) * np.sum(y - y_pred)       # Partial derivative w.r.t b
        
        # Update parameters
        m -= learning_rate * dm
        b -= learning_rate * db
        
        # Compute cost (Mean Squared Error)
        cost = (1/n) * np.sum((y - y_pred) ** 2)
        cost_history.append(cost)
    
    return m, b, cost_history

# Run Gradient Descent
learning_rate = 0.1
iterations = 500
m, b, cost_history = gradient_descent(X, y, learning_rate, iterations)

# Results
print(f"Optimized parameters: m = {m:.2f}, b = {b:.2f}")

# Plot cost history
plt.figure(figsize=(8, 6))
plt.plot(range(iterations), cost_history, label="Cost Function")
plt.title("Cost Function vs. Iterations")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.legend()
plt.show()

# Plot fitted line
plt.figure(figsize=(8, 6))
plt.scatter(X, y, label="Data", color="blue")
plt.plot(X, m * X + b, label="Fitted Line", color="red")
plt.title("Linear Regression with Gradient Descent")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Example Data
y_actual = [1, 0, 1, 0, 1]  # True labels
y_pred_probs = [0.95, 0.85, 0.78, 0.65, 0.45]  # Predicted probabilities

# Compute ROC Curve
fpr, tpr, thresholds = roc_curve(y_actual, y_pred_probs)

# Compute AUC
auc_score = roc_auc_score(y_actual, y_pred_probs)
print(f"AUC Score: {auc_score:.2f}")

# Plot ROC Curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, marker='o', label=f"ROC Curve (AUC = {auc_score:.2f})")
plt.plot([0, 1], [0, 1], linestyle="--", color="gray", label="Random Guess")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

```python
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt

# Sample 2D data points
X = np.array([[1, 2], [2, 3], [2, 2], [8, 7], [8, 8], [25, 80]])

# Applying DBSCAN with eps=3 and min_samples=2
dbscan = DBSCAN(eps=3, min_samples=2)
dbscan.fit(X)

# DBSCAN labels: -1 means noise, other numbers represent clusters
labels = dbscan.labels_

# Plot the clusters and noise points
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title('DBSCAN Clustering')
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import silhouette_score

# Load Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
data = pd.read_csv(url)

# Select relevant features
features = ['Pclass', 'Age', 'Fare', 'Sex', 'SibSp', 'Parch']
df = data[features]

# Handle missing values
df['Age'].fillna(df['Age'].mean(), inplace=True)

# Encode categorical variable 'Sex'
df['Sex'] = LabelEncoder().fit_transform(df['Sex'])

# Standardize the features
scaler = StandardScaler()
X = scaler.fit_transform(df)

# Apply DBSCAN
dbscan = DBSCAN(eps=1.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)

# Apply K-Means with 2 clusters (since we expect survival/non-survival groups)
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans.fit_predict(X)

# Evaluate performance using Silhouette Score
dbscan_silhouette = silhouette_score(X, dbscan_labels) if len(set(dbscan_labels)) > 1 else -1
kmeans_silhouette = silhouette_score(X, kmeans_labels)

print(f"DBSCAN Silhouette Score: {dbscan_silhouette:.3f}")
print(f"K-Means Silhouette Score: {kmeans_silhouette:.3f}")

# Plot clustering results
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))

# Plot DBSCAN results
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=dbscan_labels, cmap='viridis')
plt.title('DBSCAN Clustering')

# Plot K-Means results
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis')
plt.title('K-Means Clustering')

plt.show()
```
